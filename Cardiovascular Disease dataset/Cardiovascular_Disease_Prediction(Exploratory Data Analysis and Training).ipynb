# Splitting the Features and Target

```python
# creating two different datasets where:
# X = dataset without the cardio column
# Y = dataset with only the cardio column
X = df.drop(columns='cardio', axis=1)
Y = df[['cardio']]
```

```python
# printing first 10 rows of X
X.head(10)
```

```python
# printing first 10 rows of Y
Y.head(10)
```

# Scaling the Data

```python
scaler = StandardScaler()
X_standard = scaler.fit_transform(X)
```

# Splitting the Dataset into Training Data and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_standard, Y, test_size=0.2, stratify=Y, random_state=42) 
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

```python
model = GradientBoostingClassifier(n_estimators=50, learning_rate=0.2, max_depth=5, min_samples_split=10, subsample=0.8, random_state=42)
```

```python
# training the Gradient Boosting Classifier Model with Training Data
model.fit(X_train, Y_train.values.ravel())
```

# Model Evaluation

```python
# accuracy on Training Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
# printing accuracy on Training Data
print("Training Accuracy: " + str(training_data_accuracy))
```

```python
# accuracy on Test Data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
# printing accuracy on Testing Data
print("Testing Accuracy: " + str(testing_data_accuracy))
```

# Data Prediction System
